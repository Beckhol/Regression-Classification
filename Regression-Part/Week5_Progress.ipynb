```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
from sklearn import datasets, linear_model
from sklearn.model_selection import train_test_split
```

```python
#generate and filter the data
data = pd.read_csv('new_data.csv')
data['Sex'] = data['Sex'].replace('M',1)
data['Sex'] = data['Sex'].replace('F',0)
data['Side'] = data['Side'].replace('L',0)
data['Side'] = data['Side'].replace('R',1)
X = data[['Age','Sex','BMI','GX AT VO2 (mL/kg/min)','GX AT VE/VCO2','GX VO2 Max VO2 (mL/kg/min)','GX VO2 Max VO2/Pred (%)','GX VO2 Max VO2/HR (mL/beat)','GX Predicted VO2/HR (mL/beat)','GX Work Max Work (Watts)','GX Predicted Work (Watts)','PF Pre FEV1/FVC (%)','PF Pre FEV1 (L)','Side','pred mort','LOS']]
X = X.dropna(axis=0)
```

```python
#separate X,y
X1 = X[['Age','Sex','BMI','GX AT VO2 (mL/kg/min)','GX AT VE/VCO2','GX VO2 Max VO2 (mL/kg/min)','GX VO2 Max VO2/Pred (%)','GX VO2 Max VO2/HR (mL/beat)','GX Predicted VO2/HR (mL/beat)','GX Work Max Work (Watts)','GX Predicted Work (Watts)','PF Pre FEV1/FVC (%)','PF Pre FEV1 (L)','Side','pred mort']]
y = X[['LOS']]
```

```python
#Train data with 80-20
#However, if a fixed value is assigned like random_state = 
#42 then no matter how many times you execute your code the result 
#would be the same .i.e, same values in train and test datasets.
X_train, X_test, y_train, y_test = train_test_split(X1,y,test_size=0.2)

#Use Linear regression from sklearn
from sklearn.linear_model import LinearRegression
linreg = LinearRegression()
linreg.fit(X_train, y_train)
predict_data = linreg.predict(X_test)

#produce the r2_score of the model
from sklearn.metrics import r2_score
r2_score(y_test,predict_data)
```

```python
#plot the r2_score graph
#difference between the true LOS and predicted LOS
r2 = []
for i in range(100):
    X_train, X_test, y_train, y_test = train_test_split(X1,y,test_size=0.2)
    linreg = LinearRegression()
    linreg.fit(X_train, y_train)
    predict_data = linreg.predict(X_test)
    y_test = np.array(y_test)
    r2.append(r2_score(y_test,predict_data))
#     plt.plot(np.arange(45),y_test, 'bo',label='True data')
#     plt.plot(np.arange(45), predict_data, 'go',label='Predict data')
# #     for j in range(45):
#         plt.plot((j,j),(y_test[j],predict_data[j]),'r-')
#     plt.title(f"Iteration for N = {i}")
#     plt.xlabel("Testing Patient")
#     plt.ylabel("Difference")
#     plt.savefig(f"difference_{i}.png")
#     plt.legend(loc="upper left")
#     plt.show()
#     plt.close()
plt.plot(np.arange(100),r2);
plt.xlabel("Iteration Times")
plt.ylabel("r2 values for each time")
print("mean value of r2 is ",np.mean(r2))
```

```python
#plot the true LOS and predict LOS to see the difference(10 samples)
for i in range(10):
    X_train, X_test, y_train, y_test = train_test_split(X1,y,test_size=0.2)
    linreg = LinearRegression()
    linreg.fit(X_train, y_train)
    predict_data = linreg.predict(X_test)
    y_test = np.array(y_test)
    plt.plot(np.arange(45),y_test, 'bo',label='True data')
    plt.plot(np.arange(45), predict_data, 'go',label='Predict data')
    for j in range(45):
        plt.plot((j,j),(y_test[j],predict_data[j]),'r-')
    plt.title(f"Iteration for N = {i}")
    plt.xlabel("Testing Patient")
    plt.ylabel("LOS")
    plt.savefig(f"difference_{i}.png")
    plt.legend(loc="upper left")
    plt.show()
    plt.close()
```

```python
# More clear graph to see the actual difference
# using the abs
#plot the true LOS and predict LOS to see the difference(10 samples)
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import median_absolute_error
mean_absolute = []
mean_square = []
median = []
average = []
for i in range(10):
    X_train, X_test, y_train, y_test = train_test_split(X1,y,test_size=0.2)
    linreg = LinearRegression()
    linreg.fit(X_train, y_train)
    predict_data = linreg.predict(X_test)
    y_test = np.array(y_test)
    mean_absolute.append(mean_absolute_error(y_test,predict_data))
    mean_square.append(mean_squared_error(y_test,predict_data))
    median.append(median_absolute_error(y_test,predict_data))
    for j in range(45):
        plt.plot((j,j),(0,abs(y_test[j]-predict_data[j])),'r-')
    plt.title(f"Iteration for N = {i}")
    plt.xlabel("Testing Patient")
    plt.ylabel("Difference")
    plt.savefig(f"difference_{i}.png")
    plt.show()
    plt.close()
plt.plot(mean_absolute,'bo')
plt.title("Mean absolute error")
plt.ylabel("Difference between true and predict value")
plt.show()
plt.close()
plt.plot(mean_square,'bo')
plt.title("Root Mean square error")
plt.ylabel("Square Difference between true and predict value")
plt.show()
plt.close()
plt.plot(median,'bo')
plt.title("Median absolute error")
plt.ylabel("Difference between true and predict value")
plt.show()
plt.close()
```

```python
#Leave one out cross validation
from sklearn.model_selection import LeaveOneOut
from sklearn.model_selection import cross_val_score
mean_ab_error = []
mean_sq_error = []
median_error = []
X2 = np.array(X1)
y = X[['LOS']]
y = np.array(y)
loo = LeaveOneOut()
lin_reg = LinearRegression()
for train_index, test_index in loo.split(X2):
    X_train, X_test = X2[train_index], X2[test_index]
    y_train, y_test = y[train_index], y[test_index]
    lin_reg.fit(X_train,y_train)
    y_test_pred = lin_reg.predict(X_test)
    mean_ab_error.append(mean_absolute_error(y_test,y_test_pred))
    mean_sq_error.append(mean_squared_error(y_test,y_test_pred,squared=False))
    median_error.append(median_absolute_error(y_test,y_test_pred))
plt.plot(mean_ab_error,'bo')
plt.title("Mean absolute error")
plt.ylabel("Difference between true and predict value")
plt.show()
plt.close()
plt.plot(mean_sq_error,'bo')
plt.title("Root Mean square error")
plt.ylabel("Square Difference between true and predict value")
plt.show()
plt.close()
plt.plot(median_error,'bo')
plt.title("Median absolute error")
plt.ylabel("Difference between true and predict value")
plt.show()
plt.close()
```
